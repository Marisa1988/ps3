```python
##Problem 2 (5 pts):

#Write a script that:
#1 Calculate the lowest, highest and average water level observed during the time period. 
#2 Print the date and time for the lowest and highest readings.
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('/blue/bsc4452/share/Class_Files/data/CO-OPS_8727520_wl.csv')

# Remove quotes from the columns
df = df.applymap(lambda x: x.strip('"') if isinstance(x, str) else x)

# Convert the "Preliminary (ft)" column to numeric, forcing errors to NaN
df['Preliminary (ft)'] = pd.to_numeric(df['Preliminary (ft)'], errors='coerce')

# Calculate the lowest, highest, and average values
lowest_value = df['Preliminary (ft)'].min()
highest_value = df['Preliminary (ft)'].max()
average_value = df['Preliminary (ft)'].mean()

# Find the rows with the lowest and highest values
lowest_row = df.loc[df['Preliminary (ft)'].idxmin()]
highest_row = df.loc[df['Preliminary (ft)'].idxmax()]

# Extract the date and time for the lowest and highest values
lowest_date = lowest_row['Date']
lowest_time = lowest_row['Time (GMT)']
highest_date = highest_row['Date']
highest_time = highest_row['Time (GMT)']

# Report the lowest, highest, and average values and the corresponding Date and Time (GMT)
print(f"The lowest water level was {lowest_value} ft, observed on {lowest_date} at {lowest_time} GMT.")
print(f"The highest water level was {highest_value} ft, observed on {highest_date} at {highest_time} GMT.")
print(f"The average water level was {average_value:.2f} ft.")
```

```python
##Problem 3 (5 pts):

#Write a script that:
#1 calculates the fastest rise in water level per 6-minute period between measurements
#2 reports the date and time that was observed 
#3 and the change in water level from the previous recording.
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('/blue/bsc4452/share/Class_Files/data/CO-OPS_8727520_wl.csv')

# Remove quotes from the columns
df = df.applymap(lambda x: x.strip('"') if isinstance(x, str) else x)

# Convert the "Preliminary (ft)" column to numeric, forcing errors to NaN
df['Preliminary (ft)'] = pd.to_numeric(df['Preliminary (ft)'], errors='coerce')

# Calculate the differences between consecutive rows in the "Preliminary (ft)" column
df['Change'] = df['Preliminary (ft)'].diff().abs()

# Find the row with the biggest change
max_change_row = df.loc[df['Change'].idxmax()]

# Extract the date, time, and the biggest change
date_value = max_change_row['Date']
time_value = max_change_row['Time (GMT)']
biggest_change = round(max_change_row['Change'], 2)

# Report the biggest change and the corresponding Date and Time (GMT)
print(f"The fastest rise in water level per 6-minute period was {biggest_change}, observed on {date_value} at {time_value} GMT.")
```

```python
##Problem 4 (5 pts):

#Imagine that the file is providing live readings of the water level. 
#Write a script that print a line of text with a warning if any of these events occur:
#1 The water level increases more than 0.25 since the previous recording.
#2 The water level is over 5.0.
#3 No reading is received at a time point.
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('/blue/bsc4452/share/Class_Files/data/CO-OPS_8727520_wl.csv')

# Remove quotes from the columns
df = df.applymap(lambda x: x.strip('"') if isinstance(x, str) else x)

# Convert the "Preliminary (ft)" column to numeric, forcing errors to NaN
df['Preliminary (ft)'] = pd.to_numeric(df['Preliminary (ft)'], errors='coerce')

# Iterate through the DataFrame to check for the specified events
for i in range(1, len(df)):
    date = df.at[i, 'Date']
    time = df.at[i, 'Time (GMT)']
    current_value = df.at[i, 'Preliminary (ft)']
    previous_value = df.at[i-1, 'Preliminary (ft)']

    # Check if no value is received at a time point
    if pd.isna(current_value):
        print(f"Warning: No value received at {date} {time} GMT.")
        continue

    # Check if the preliminary (ft) value increases more than 0.25 since the previous value
    if not pd.isna(previous_value) and (current_value - previous_value) > 0.25:
        print(f"Warning: The water level increased more than 0.25 ft at {date} {time} GMT.")

    # Check if the preliminary (ft) value is over 5.0
    if current_value > 5.0:
        print(f"Warning: The water level is over 5.0 ft at {date} {time} GMT.")
```
